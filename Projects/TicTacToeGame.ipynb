```python
import random
print('Format of the Game !!')
demo = """

  1 | 2 | 3
----|---|-----
  5 | 4 | 6
----|---|------
  7 | 8 | 9
 """
print(demo)

sign = [' '] * 10

def board():
    board_str = f'''
      {sign[1]} | {sign[2]} | {sign[3]}
    ----|---|-----
      {sign[4]} | {sign[5]} | {sign[6]}
    ----|---|------
      {sign[7]} | {sign[8]} | {sign[9]}

     '''
    print(board_str)

def Game():
    name = input('Enter your name: ').title()
    choice = input('Enter your choice \'X\' OR \'0\': ').upper()

    while choice not in ['X', '0']:
        print("Invalid choice. Please enter 'X' or '0'.")
        choice = input('Enter your choice \'X\' OR \'0\': ')

    player2 = 'Computer'

    if choice == 'X':
        computer_choice = '0'
    else:
        computer_choice = 'X'

    print('-----------------------------------')
    print('Player 1 is Human !!')
    print('Name:', name)
    print('Input choice:', choice)
    print('-----------------------------------')
    print('Player 2 is computer !!')
    print('Name:', player2)
    print('Input choice:', computer_choice)
    print('-----------------------------------')

    first_turn = random.choice([name, player2])
    print(f"Randomly deciding the first turn...\n")
    print(f"******{first_turn} will go first!******")
    board()

    for i in range(1, 10):
        if first_turn == name:
            position = int(input('Enter Position (1-9) : '))
            while sign[position] != ' ':
                print('Position is already filled. Enter again !!')
                position = int(input('Enter Position (1-9) : '))
            sign[position] = choice
        else:
            print('Computer Turn')
            num = random.randint(1, 9)
            while sign[num] != ' ':
                num = random.randint(1, 9)
            sign[num] = computer_choice

        board()

        if check_win(sign, choice):
            print(f"{name} wins!")
            break
        elif check_win(sign, computer_choice):
            print("Computer wins!")
            break
        elif i == 9:
            print("It's a draw!")

        first_turn = player2 if first_turn == name else name

def check_win(board, player):
    return ((board[1] == board[2] == board[3] == player) or
            (board[4] == board[5] == board[6] == player) or
            (board[7] == board[8] == board[9] == player) or
            (board[1] == board[4] == board[7] == player) or
            (board[2] == board[5] == board[8] == player) or
            (board[3] == board[6] == board[9] == player) or
            (board[1] == board[5] == board[9] == player) or
            (board[3] == board[5] == board[7] == player))

Game()

re = input('You want to play again ? Yes > \'Y\' or No > \'N\' : ').upper()
if re =='Y':
    sign = [' '] * 10
    Game()
elif re =='N':
    print('Thank you for playing the game')
else:
    print('Invalid Input !!')
```
