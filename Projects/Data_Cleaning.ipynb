```python
#to find percentage of missing value
null_val = df.isnull().sum() / df.shape[0] *100
null_val
```

```python
# see how many columns has more than 17% missing value
drop_columns = null_val[null_val >17].keys()
drop_columns
```

```python
# now drop columns having greater tham 17% mising value
df2_drop_column = df.drop(columns = drop_columns)
```

```python
df2_drop_column.shape
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df2_drop_column.isnull())
```

```python
df3_drop_rows = df2_drop_column.dropna()
df3_drop_rows
```

```python
cat_col = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
       'PavedDrive', 'SaleType', 'SaleCondition']
```

```python
def cat_val(var):
  return pd.concat([df[var].value_counts() / df.shape[0] *100, df3_drop_rows[var].value_counts() /df3_drop_rows.shape[0]  * 100], axis=1, keys=[var+'_Original', var+'_Cleaned'])
for i in cat_col:
  result = cat_val(i)
  print(result)
  print('\n\n\n')
```
