```python
# 5) find mean salary of people with age above 50
x =df.Wage[df['Age']>50].mean()
print(x)
```

```python
# 6) find mean salary of people with age below 20
y=df.Wage[df['Age']<20].mean()
print(y)
```

```python
# 7) find which education category has highest wage
dictonary_of_mean_wage={}

for k in df.Education.value_counts().keys():
  dictonary_of_mean_wage[k]=df.Wage[df['Education']==k].mean()

print(dictonary_of_mean_wage)
# get the key from dictonary which has highest value
max_key = max(dictonary_of_mean_wage, key=dictonary_of_mean_wage.get)
print("Key with the highest value:", max_key)


```

```python
# find education and age for highest and lowest wage

print(df[df['Wage']==df['Wage'].min()])
print("===================================================")
print(df[df['Wage']==df['Wage'].max()])
```
