```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.scatter(df.area, df.price,color='red',marker='+')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Housing Price Analysis')
```

```python
df.area, df.price
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
reg.fit(new_df,price)
```

```python
reg.predict([[3300]]) # 628715.75342466
```

```python
'''
y = mx + c
y = dependent variable
m = slope
x = independent variable
c = intercept
'''
```

```python
reg.coef_ # 135.78767123
```

```python
reg.intercept_ # 180616.43835616432
```

```python
(135.78767123*3300)+180616.43835616432 # [628715.75342466]
```

```python
new_rows = []
for _ in range(26):
    new_row = {
        'area': np.random.randint(2000, 5000),  # Random 'area' between 2000 and 5000
        'price': np.random.randint(500000, 800000)  # Random 'price' between 500000 and 800000
    }
    new_rows.append(new_row)

# # Concatenate the original DataFrame and the new rows
df = pd.concat([df, pd.DataFrame(new_rows)], ignore_index=True)

# # Print the updated DataFrame
df





```
