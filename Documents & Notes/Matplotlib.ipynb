**Line Plot**

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,11)
y = 2*x
plt.plot(x,y, color='g', linestyle=':', linewidth=5)
plt.title('Line Plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1, 11)
y1 = 2 * x
y2 = 3*x
plt.plot(x,y1, color='r', linestyle=":", linewidth=5)
plt.plot(x, y2, color='y', linewidth=5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Two Line in Same Plot')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x

plt.subplot(1,2,1)
plt.plot(x,y1, color ='g', linestyle=':', linewidth=5)

plt.subplot(1,2,2)
plt.plot(x,y2, color='r', linewidth=5)

plt.title('Two line')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()

```

**Bar Plot**

```python
student = {'Ram':98, 'Shyam':78, 'Geeta':90, 'Rita':65 }
name = list(student.keys())
marks = list(student.values())
plt.bar(name, marks)
plt.title('Student Result')
plt.xlabel('Student Name')
plt.ylabel('Marks Obtained')
plt.show()
```

**Horizontal Bar**

```python
student = {'Ram':98, 'Shyam':78, 'Geeta':90, 'Rita':65 }
name = list(student.keys())
marks = list(student.values())
plt.barh(name, marks, color='r')
plt.title('Student Result')
plt.xlabel('Student Name')
plt.ylabel('Marks Obtained')
plt.show()
```

# **Scatter-Plot**

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
a = [4, 5, 2, 7, 8,3 ,9 , 4, 8]
b =[6,2, 7, 5, 9, 2, 4, 7, 5]
plt.scatter(x,a, marker ='*', c='g', s=100)
plt.scatter(x,b, marker='.', c='b', s=100)
plt.title('Scatter Plot')
plt.show()
```
