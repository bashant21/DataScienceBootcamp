```python
#Make marks half

def half(x):
  return x/2
```

```python
df['Marks'].apply(half)
```

```python
#New column named Half Marks added
df['Half_Marks'] = df['Marks'].apply(half)
df
```

```python
#Apply lambda function
df['Marks'].apply(lambda x: x/2)
```

```python
#length of string available in name column
df['Name'].apply(len)
```

```python
#Map
df['Male_Female'] = df['Gender'].map({'Male':1, 'Female':0})
df
```

```python
#drop columns
df.drop('Male_Female', axis=1)
```

```python
#drop Multiple columns
df.drop(['Male_Female', 'Half_Marks'], axis=1, inplace=True)
```
